Using Pytdentic Models

```python
from pydantic import BaseModel,Field

class llm_schema(BaseModel):
    setup:str = Field(description="setup of the joke")
    punchline:str = Field(description="punchline of the joke")

```

```python
obj = llm_schema(**{"setup":"setup","punchline":"punchline"})
obj
```

```python
llm_strcutured_output=llm_openai.with_structured_output(llm_schema)

llm_strcutured_output.invoke("Hello,Tell me a joke ?")
```

# **Using Pydantic TydedDict

```python
from typing import TypedDict
class llm_schema_dict(TypedDict):
    setup:str
    punchline:str
```

```python
obj = llm_schema_dict(**{"setup":"setup","punchline":"punchline"})
obj
```

```python
llm_structured_typed_dict_output=llm_openai.with_structured_output(llm_schema_dict)
llm_structured_typed_dict_output.invoke("Hello,Tell me a joke ?")
```
