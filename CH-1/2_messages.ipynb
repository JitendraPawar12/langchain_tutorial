```python
from langchain_core.messages import HumanMessage,SystemMessage,AIMessage

my_messages = [
SystemMessage(content="You are a gen-z assistant who always answers in fun way"),
HumanMessage(content="Tell me a fun fact")
]

from langchain_openai import ChatOpenAI

llm_openai=ChatOpenAI(model="gpt-5-mini", temperature=0)
llm_openai.invoke(my_messages).content
```

```python
from langchain_core.prompts import  PromptTemplate


dynamic_prompt=PromptTemplate.from_template("Fun fact about {topic}?")
ready_prompt=dynamic_prompt.invoke({"topic": user_input})
llm_openai.invoke(ready_prompt).content
```

```python
from langchain_core.prompts import ChatPromptTemplate

user_input=input("Enter a topic for a fun fact: ")
user_tone=input("Enter the tone you want the answer in (e.g., fun, formal, sarcastic): ")
promts=ChatPromptTemplate.from_messages([
    ("system","You are a gen-z assistant who always answers in {tone} way"),
    ("user","Tell me a fun fact about {topic}")
])

ready_prompt=promts.invoke({"topic": user_input, "tone": user_tone})
llm_openai.invoke(ready_prompt).content


```
