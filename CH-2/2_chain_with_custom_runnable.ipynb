```python
from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv

load_dotenv()

if os.environ.get("OPENAI_API_KEY"):
    print("OPENAI_API KEY is set")
else:
    raise ValueError("OPENAI_API_KEY is not set in environment variables")

from langchain_core.prompts import  ChatPromptTemplate,PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_core.runnables import RunnableLambda, RunnableSequence

llm_openai=ChatOpenAI(model="gpt-5-mini", temperature=0.9, max_tokens=1000)
llm_openai.invoke("what is the capital of italy?")
```

Chain with Custom Runnable

```python
#Task -1 

prompt_template = ChatPromptTemplate.from_messages(
    [("system","You are a helpful assistant that prvodes information?"),
     ("human","{input}")])

print(prompt_template)     
```

```python
#Task -2

llm_openai = ChatOpenAI(model="gpt-5-mini", temperature=0.9, max_tokens=1000)
```

```python
#Task -3

str_parser=StrOutputParser()
```

```python
#task -4

def dictionary_maker(input:str)->dict:
    return {"input": input}

dictionary_maker_runnable=RunnableLambda(dictionary_maker)
```

```python
#task -5
prompt_post=ChatPromptTemplate.from_messages(
    messages=[
        ("system","You are a social media post generator that creates engaging posts based on the given input."),
        ("human","{input}")  
    ])
```

```python
#Task -6
llm_openai = ChatOpenAI(model="gpt-5-mini", temperature=0.9, max_tokens=1000)
```

```python
#Task -7

str_parser=StrOutputParser()
```

```python
#Task -8
chain = prompt_template | llm_openai | str_parser | dictionary_maker_runnable | prompt_post | llm_openai | str_parser

# To see intermediate outputs, test each step
input_text = "What is the capital of France?"
print("Initial input:", input_text)

# Step 1: Through prompt_template, llm_openai
step1 = (prompt_template | llm_openai).invoke(input_text)
print("\nAfter prompt + LLM :", step1)

# Step 1: Through prompt_template, llm_openai, str_parser
step1 = (prompt_template | llm_openai | str_parser).invoke(input_text)
print("\nAfter prompt + LLM + parser:", step1)

# Step 2: Through dictionary_maker_runnable
step2 = dictionary_maker_runnable.invoke(step1)
print("\nAfter RunnableLambda (dictionary_maker):", step2)

# Full chain result
result = chain.invoke(input_text)
print("\nFinal chain output:", result)
```
